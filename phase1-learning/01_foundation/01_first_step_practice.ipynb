```python
def load_ecommerce_data(month):
    return (spark.read.format("csv")
            .option("header", "true")
            .option("inferSchema", "true")
            .load(f"/Volumes/workspace/ecommerce/ecommerce_data/2019-{month}.csv"))

def get_valid_month():
    while True:
        month = input("Enter the month to load the data from: ").capitalize()
        if month in ("Oct", "Nov"):
            return month
        print(f"Invalid month: {month}. Please enter either Oct or Nov.")

month: str = get_valid_month()
print(f"Loading data for {month}...")

# Load the data
events = load_ecommerce_data(month)

# Verify it's working
print(f"âœ… Ready to go! Loaded {events.count():,} events")
display(events.limit(5))
```
